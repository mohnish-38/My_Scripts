## Here we grab any image from web and save to our computer

```python
import requests as r
#PIL stands for python imaging library 
from PIL import Image  as i
from PIL import ImageFilter as f
from io import BytesIO as b

url=input()

response=r.get(url)

img=i.open(b(response.content))
img #we directly see image without saving to computer
```

## we can save image to our desktop ,check this out

```python
url=input()

response=r.get(url)

img=i.open(b(response.content))
img=img.save("image.jpg")
```
